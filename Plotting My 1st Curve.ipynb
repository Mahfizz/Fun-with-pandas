```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import matplotlib.font_manager
from IPython.core.display import HTML

def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(font=fontname)

code = "\n".join([make_html(font) for font in sorted(set([f.name for f in matplotlib.font_manager.fontManager.ttflist]))])

HTML("<div style='column-count: 2;'>{}</div>".format(code))
```

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]
plt.figure(figsize=(4,3),dpi=300)
plt.plot(x,y,label=('Y=2X'),color='red',linewidth=2,marker='.',linestyle='-',markersize=10,markeredgecolor='yellow')
plt.title('My First Plot', fontdict={'fontname' : 'MS Gothic', 'fontsize' : 10})
m=np.arange(0,4.5,0.5)
n=m**2
plt.plot(m[0:4],n[0:4],'g.-')
plt.plot(m[3:],n[3:],'b.--',label='Y=X^2')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])
plt.legend()
plt.savefig('yoo.png', dpi=300)
plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [1,3,2]
bars= plt.bar(labels,values,color='yellow')
patterns=['/','*','o']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
plt.show()
```
